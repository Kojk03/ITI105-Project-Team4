A modification have been done on the NearPy library. The original library includes the original vector (which is normalized) in the hashtable. Due to the fact for recommender systems where each vector consist of a large number of columns (in this dataset >45k columns. Even after converting to a dense vector, it still causes memory issues), the resulting hashtable is large enough to cause memory issues (at least with my PC). The library is modified so that the vectors are not stored. The drawback is that distances cannot be calculated in query function. That have also been removed. The whole list of candidates will be returned. The vectors and therefore the distance and the rankings will be retrieved/calculated afterwards.
